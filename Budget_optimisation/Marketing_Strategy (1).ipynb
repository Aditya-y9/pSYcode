```python
import pandas as pd

# Step 1: Read ROI Data from CSV
roi_data = pd.read_csv('consolidated_summary.csv')

# Step 2: Define Campaigns of Interest
variables_of_interest = [
    'Reel_0_8_Story_0_2',
    'Campaign_Story',
    'Campaign_Reel',
    'Post_0_5_Story_0_5',
    'Post_0_4_Reel_0_6'
]

# Step 3: Preprocess 'Return on Investment' column and Calculate Overall ROI for Each Campaign of Interest
campaign_roi = {}
for variable in variables_of_interest:
    # Convert 'Return on Investment' column to string and then preprocess
    roi_data['Return on Investment'] = roi_data['Return on Investment'].astype(str).str.replace('$', '').astype(float)
    campaign_roi[variable] = roi_data[roi_data['Variable'] == variable]['Return on Investment'].mean()

# Step 4: Set Total Budget
total_budget = float(input("Enter total budget: "))

# Step 5: Allocate Budget Proportionally Based on ROI
total_roi = sum(campaign_roi.values())
budget_allocation = {}

for campaign, roi in campaign_roi.items():
    budget_allocation[campaign] = total_budget * (roi / total_roi)

# Step 6: Save Budget Allocation to CSV File
budget_allocation_df = pd.DataFrame(budget_allocation.items(), columns=['Campaign', 'Budget Allocation'])
budget_allocation_df.to_csv('budget_allocation.csv', index=False)

# Step 7: Output Message
print("Budget allocation saved to 'budget_allocation.csv' file.")
```
